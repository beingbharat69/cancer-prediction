# 1) Importing Libraries:

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2) Loading the dataset:

```python
df= pd.read_csv('data.csv')#view our dataset using head():
df.head(10)
```

# 3) Analysing the data:

```python
# count the number of rows and columns in dataset:
df.shape
```

```python
sns.pairplot(df,hue = 'diagnosis', palette= 'coolwarm', vars = ['radius_mean', 'texture_mean', 'perimeter_mean','area_mean','smoothness_mean'])
```

```python
# count the number of empty values in each columns:
df.isna().sum()# drop the columns with all the missing values:
df = df.dropna(axis = 1)
df.shape# Get the count of the number of Malignant(M) or Benign(B) cells
df['diagnosis'].value_counts()
```

```python
df.describe()
```

```python
# visualize the count:
sns.countplot(df['diagnosis'], label = 'count')
```

```python
# Rename the diagnosis data to labels:
df = df.rename(columns = {'diagnosis' : 'label'})
print(df.dtypes)
```

```python
# define the dependent variable that need to predict(label)
y = df['label'].values
print(np.unique(y))
```

# 4) Encoding Categorical Data:

```python
# Encoding categorical data from text(B and M) to integers (0 and 1)
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(y) # M = 1 and B = 0
print(np.unique(Y))
# maligant ype as 1 benign as 0
```

# 5)Defining the independent variables, Dropping labels and IDs, and normalizing other data:

```python

X  = df.drop(labels=['label','id'],axis = 1)
#scale / normalize the values to bring them into similar range:
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
print(X)
```

# 6) Splitting Our data:

```python
# Split data into training and testing data to verify accuracy after fitting the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.25, random_state=42)
print('Shape of training data is: ', x_train.shape)
print('Shape of testing data is: ', x_test.shape)
```

```python
print ('Number of training samples input is', x_train.shape)
print ('Number of testing samples input is', x_test.shape)
print ('Number of training samples output is', y_train.shape)
print ('Number of testing samples output is', y_test.shape)
```

# 7) Creating Model:

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
```

```python
model = Sequential()
model.add(Dense(128, input_dim=30, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64,activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

# 8) Compile and fit ml model to our training data:

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam' , metrics = ['accuracy'])
model.summary()
```

```python
# fit with no early stopping or other callbacks:
history = model.fit(x_train,y_train,verbose = 1,epochs = 100, batch_size = 64,validation_data = (x_test,y_test))
```

# 9) Visualizing our training accuracy and validation accuracy:

```python
# plot the training and validation accuracy and loss at each epochs:
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1,len(loss)+1)
plt.plot(epochs,loss,'y',label = 'Training loss')
plt.plot(epochs,val_loss,'r',label = 'Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.plot(epochs,acc,'y',label = 'Training acc')
plt.plot(epochs,val_acc,'r',label = 'Validation acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

# 10) Prediction and Visualizing our model accuracy on test data:

```python
# Predicting the Test set results:
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
# Making the Confusion Matrix:
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot = True)
```
